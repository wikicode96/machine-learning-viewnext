## Importar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generar datos para el eje x

```python
x = np.linspace(0, 2 * np.pi, 100)  # 100 puntos entre 0 y 2π
```

## Calcular los valores de y usando la función seno

```python
y = np.sin(x)
```

## Crear la gráfica

```python
plt.plot(x, y, '.', label='sin(x)')

# Añadir título y etiquetas
plt.title("Gráfica de la función seno")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Añadir una cuadrícula para mejor visualización
plt.grid(True)

# Mostrar la leyenda
plt.legend()
```
